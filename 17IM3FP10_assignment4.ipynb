```python
f = open("test.txt",'r') 
str=f.read()
str.split(".")
i=0
l=[]
for line in str.split("."):
    if i%2==0:
        l.append(line)
    i=i+1
new_str = ''
for word in l:
    new_str = new_str + ' ' + word
new_str
f_new=open("test_new.txt",'w')
f_new.write(new_str)
```

```python
def remove_newlines(f):
    f_list = open(f).readlines()
    return [s.rstrip('\n') for s in f_list]

print(remove_newlines("test.txt"))
```
